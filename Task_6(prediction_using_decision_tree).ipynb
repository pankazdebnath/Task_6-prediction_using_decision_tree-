Reading data from drive

```python
#read the csv file from base directory with pandas
iris_df = pd.read_csv("Iris.csv")
iris_df.head()
```

Preparing the data for classification

```python
#splitting the features and the classes for creating classifier
X = iris_df.drop(['Species', 'Id'], axis = 1)
y = iris_df['Species']
print("Train features: ")
print(X.head())
print("Classification: ")
print(y.head())
```

Creating and training the Decision Tree Classifier model

```python
#Defining the Decision Tree Algorithm
from sklearn.tree import DecisionTreeClassifier

algo = DecisionTreeClassifier()    #defining algorithm as model
dt_clf = algo.fit(X, y)            #fitting the data into the model

print("Decision Tree Classifier Created")
```

Visualization of Decision Tree according to the data(iris)

```python
from sklearn import tree            #for plotting the tree
import matplotlib.pyplot as plt     #for resizing and visualization
plt.figure(figsize=(20,15),
           facecolor= '#B6F5C2')
tree.plot_tree(dt_clf,
               filled = True,
               rounded = True,
               node_ids= True,
               feature_names = X.columns,
               label = "all",
               impurity = True)
plt.title("Decision Tree trained by the data features", color='#B00B79')
plt.show()
```
